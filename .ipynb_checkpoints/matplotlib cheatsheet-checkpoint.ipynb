# Imports

```python
import matplotlib.pyplot as plt
import numpy as np

# Inline rendering
%matplotlib inline

# widget - jupyter lab and notebook > 7
# %matlablib widget

# Notebook
# %matlablib notebook
```

# chapter 1 - basics

## simple plot

```python
# Prepare the data
x = np.linspace(0,10*np.pi, 1000)
y = np.sin(x)
```

```python
# plot
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

## scatter plot

```python
# Prepare the data

x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)
```

```python
# Plot

fig,ax = plt.subplots()
ax.scatter(x,y)
```

## barplot

```python
# Prepare the data

x = np.arange(10)
y = np.random.uniform(1,10,10)
```

```python
# Bar plot

fig, ax = plt.subplots()
ax.bar(x,y)
```

## Imshow

```python
z = np.random.uniform(0,1,(8,8))

fig, ax = plt.subplots()

ax.imshow(z)
```

## Contourf

```python
# Prepare the data

z = np.random.uniform(0,1,(8,8))
```

```python
# Contourf plot

fig, ax = plt.subplots()
ax.contourf(z)
```

## Pie plot

```python
# Prepare the data

z = np.random.uniform(0,1,4)
```

```python
fig, ax = plt.subplots()
ax.pie(z)
```

## Hist plot

```python
# Prepare the data

z = np.random.normal(0,1,100)
```

```python
# Plot

fig, ax = plt.subplots()
ax.hist(z)
```

## Error plot

```python
# Prepare the data

x = np.arange(5)
y = np.random.uniform(0,1,5)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x,y, y/4)
```

## Box plot

```python
# Prepare the data

z = np.random.normal(0,1,(100,3))
```

```python
# plot

fig, ax = plt.subplots()
ax.boxplot(z)
```
