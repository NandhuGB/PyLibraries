#### Manipulation of Tensors (tensor operations)

```python
# Addition

## with constant value

tensor1 = torch.tensor([1,2,3])

const  = 10

print("Addition operation with constant")

### using symbol +

print(tensor1 + const)

### using torch method

print(torch.add(tensor1, const))

print("Addition operation bw two tensor")
## with another tensor

tensor2 = torch.tensor([2,4,6])

### using symbol +

print(tensor1+ tensor2)

### using torch method

print(torch.add(tensor1, tensor2))
```

```python
# Substraction

tensor1 = torch.tensor([10, 20, 30])

## with constant

const  =  5

print("Substraction with constant")

### using symbol -

print(tensor1 - const)

### using torch method

print(torch.sub(tensor1 ,  const))

## with another tensor

tensor2 = torch.tensor([5,6,7])

print("Substraction with another tensor")

### using symbol -

print(tensor1 - tensor2)

### using torch method

print(torch.sub(tensor1, tensor2))
```

```python
# Multiplication - two ways of multiplication (normal multiplication / matrix multiplication)

tensor1 = torch.tensor([1,2,3])

## with constant (scalar)

scalar = 10

print("Multiplication with scalar")

### using symbol *

print(tensor1 * scalar)

### using torch method

print(torch.mul(tensor1, scalar))

## with another tensor

print("Multiplication with another tensor")

tensor2 = torch.tensor([2, 4, 6])

### using symbol *

print(tensor1 * tensor2)

### using torch method

print(torch.mul(tensor1, tensor2))
```

```python
# Matrix multiplication (dot product)

tensor1 = torch.tensor([[1,2,3],[4,5,7],[9,8,1]])

transformation = torch.tensor([[1,1,1],[2,2,2],[3,3,3]])

print("Matrix Multiplication-  where order of tensor matters")

## using torch method

print(torch.matmul(tensor1, transformation))

## order 2

print(torch.mm(transformation, tensor1)) ## torch.mm is alias of matmul

print("Its not equal because matrix multiplication is associative, means A.B != B.A")
```

### Tensor aggregation

```python
# Creating the tensor


tensor = torch.arange(0,100,1)

## min - tensor.min() or torch.min(tensor)

print(f"The minimum value of tensor: {torch.min(tensor)}")

## max - tensor.max() or torch.max(tensor)

print(f"The maximun value of tensor: {torch.max(tensor)}")

## mean - tensor.mean() or torch.mean(tensor)

print(f"The mean value of tensor: {torch.mean(tensor.type(torch.float32))}")


## sum -  tensor.sum() or torch.sum(tensor)

print(f"The summation of tensor: {torch.sum(tensor)}")
```

```python
# Positional min, max (argmin, argmax)

## position of minimun value - tensor.argmin() or torch.argmin(tensor)

print(f"The index of minimum value in the tensor: {torch.argmin(tensor)}")

## position of maximum value -  tensor.argmax() or torch.argmax(tensor)

print(f"The index of the maximum value in the tensor: {torch.argmax(tensor)}")
```

```python
# Reshaping and view, stacking, squeezing and unsqueezing tensors


tensor = torch.arange(0, 12)

print(f"The tensor: {tensor} with shape and length:{tensor.shape}, {len(tensor)}")

## Reshaping

tensor_reshaped = tensor.reshape(4,3)

print(tensor_reshaped)
```

```python
## View

### view is copy of the tensor with customised shape

z = tensor_reshaped.view(3,4) # think like `view` as pointer of some tensor

print(z)

### z and tensor_reshaped has same memory. means changes `z` changes `tensor_reshaped` and viceversa

z[0, 1] = 10

print("after the z[0,1] = 10")

print(f"z: {z}")

print(f"tensor_reshaped: {tensor_reshaped}")
```

```python
# Stacking

print(f"Stacked tensor: {tensor_reshaped}, with shape {tensor_reshaped.shape}")

## dim= 0 (vertical stack) like adding rows together axis = 0

stacked = torch.stack((tensor_reshaped, z.reshape(4,3)), dim= 0)

print(f"vertically stacked : {stacked}, and shape: {stacked.shape}")


## dim = 1 (horizondal stack) like adding new columns axis = 1

stacked = torch.stack((tensor_reshaped, z.reshape(4,3)), dim = 1)

print(f"Horizodally stacked: {stacked}, and shape: {stacked.shape}")
```

```python
# Squeeze

tensor = torch.arange(0,10).reshape(1,10)
## removes all the single dimension from target tensor

print(f"tensor used: {tensor}, shape : {tensor.shape}")


tensor_squeezed = torch.squeeze(tensor)

print(f"The squeezed tensor: {tensor_squeezed}, Shape: {tensor_squeezed.shape}")


## unsqueezing

tensor_unsqueezed = torch.unsqueeze(tensor_squeezed, dim = 0)

print(f"The Unsqueezed tensor: {tensor_unsqueezed}, Shape: {tensor_unsqueezed.shape}")
```

```python
# Permute

tensor = torch.rand(size = (1,2,3))

print(f"Tensor used:{tensor}, shape: {tensor.shape} ")

## Rearrange tensor axis order - premuted tensor and output tensor shares the same memory

tensor_premuted = tensor.permute(2,0,1) # or torch.premute(tensor, dims)

print(f"Tensor premuted to 2,0,1 : {tensor_premuted}, shape: {tensor_premuted.shape}")
```

### Pytorch and numpy

```python
# we can convert torch tensor into numpy array

## tensor

tensor =  torch.arange(1,10).reshape(1,3,3).type(dtype = torch.int32)

## converting into numpy array

numpy_array = tensor.numpy()

print(f"Tensor used: {tensor} with {tensor.dtype}") # data types --> float 64

print(f"Numpy array: {numpy_array} with {numpy_array.dtype}") # data types --> float 64
```

```python
# creating tensor from numpy array

array = np.arange(1, 10).reshape(3,3)

print(f"\nThe array used: {array}, \tdtype:{array.dtype} ")

## convertion

tensor = torch.from_numpy(array)


print(f"\nThe converted tensor: {tensor}, \tdtype:{tensor.dtype}")
```

### Randomness

```python
# torch.manual_seed(seed = RANDOM_SEED)


## why - neural networks initial weights are fixed randomly. then after the experiments, how to reproduce the same results with random weights. we can control randomness


### example
print("\nDefault randomness in  action")
print(f"\nTensor 1: {torch.rand(2,2)}")
print(f"\nTensor 2: {torch.rand(2,2)}")

### setting controlled randomness

print(f"\nControlled randomness in action with seed 64")

torch.random.manual_seed(seed=64) # this value reset for every random call
print(f"\nTensor 1: {torch.rand(2,2)}")

torch.random.manual_seed(seed=64)
print(f"\nTensor 2: {torch.rand(2,2)}")

#### changing the seed

print(f"\nControlled randomness in action with seed 46")

torch.random.manual_seed(seed=46)
print(f"\nTensor 1: {torch.rand(2,2)}")

torch.random.manual_seed(seed=46)
print(f"\nTensor 2: {torch.rand(2,2)}")
```
