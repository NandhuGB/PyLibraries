# Chapter 2 - Tweaks

```python
# Prepare the data

x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
# Normal plot

fig, ax = plt.subplots(figsize = (4,2))
ax.plot(x,y)
```

```python
# Changing the color

fig, ax = plt.subplots(figsize = (4,2))
ax.plot(x,y, color = "black")
```

```python
# Line style

fig, ax = plt.subplots(figsize = (4,2))
ax.plot(x,y, linestyle = "--", color = "green")
```

```python
# Changing the linewidth

fig, ax = plt.subplots(figsize = (4,2))
ax.plot(x,y, color = "red", linestyle = "--", linewidth = 2.5)
```

```python
# Marker 

fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x,y, color = "red", marker = "o")
```

# Chapter 3 - Organise

```python
# Prepare the data

x = np.linspace(0,10, 100)
y1, y2 = np.sin(x), np.cos(x)
```

## Adding multiple plot in one axis

```python
fig, ax = plt.subplots(figsize = (10,5))

ax.plot(x, y1, x, y2)
```

## subplots

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (10,3))

ax1.plot(x, y1, color = "red")
ax2.plot(x, y2, linestyle = "--", color = "green")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (4, 5))
ax1.plot(y1, x, color = "green")
ax2.plot(y2, x, linestyle = "--")
```

## Labelling

```python
# Prepare the data

x = np.linspace(0,10,100)
y = np.sin(x)
```

```python

fig, ax = plt.subplots(figsize = (10,2))
ax.plot(x,y)
ax.set_title("Sin plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

# Figures, axes & spines

```python
# Subplots - set_facecolor

fig, axes = plt.subplots(3,3, figsize = (10,8))
axes[0,0].set_facecolor("#d2a9f2")
axes[2,2].set_facecolor("#4a23df")
```

```python
# changing grid space

fig = plt.figure(figsize = (12,8))

gs = fig.add_gridspec(3,3) # dividing figure into 3x3

ax1 = fig.add_subplot(gs[0,:]) # plotting the line plot with entire row 0
ax1.plot(x,y, color = "green", linestyle = "--")

ax2 = fig.add_subplot(gs[1, : 2]) # plotting the scatter plot with row 1 until subplot 2
ax2.scatter(np.random.uniform(0,10,100), np.random.uniform(0,10,100))

ax3 = fig.add_subplot(gs[1, 2])
ax3.plot(x,y)
plt.show()
```

```python
# Spines

fig, ax = plt.subplots()

ax.spines["top"].set_color("None")
ax.spines["right"].set_color("None")
```

## Ticks and labels

```python
plt.plot([1,2,3,4,5])
plt.minorticks_on()
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF
plt.plot([1,2,3,4,5])
ax.xaxis.set_minor_locator(ML(0.2))
ax.xaxis.set_minor_formatter(SF())
plt.gca().tick_params(axis = "x", which = "minor", rotation = 90)
plt.minorticks_on()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FuncFormatter

fig, ax = plt.subplots()

# Set locators
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.2))

# Set formatters
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.1f} units"))

# Customize ticks
ax.tick_params(axis="x", which="major", length=8, rotation=45)
ax.tick_params(axis="x", which="minor", length=4, color="blue")

plt.show()
```
